1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self,maximum_number,count=0):
        self.maximum_number=maximum_number
        self.count=count
    
    def increment(self):
        if self.count >= self.maximum_number:
            raise ValueError("Error, counter has reached its maximum_value",self.maximum_number)
        else:
            self.count+=1
        
    def reset(self):
        self.count=0
    
    def __repr__(self):
        return f"counter(count={self.count}, maximum={self.maximum_number})"
    

```

```python
class Counter:
    def __init__(self,maximum_number,count=0):
        if count > maximum_number:
            raise ValueError("Initial value can't exceed the maximum value")
        self.__maximum_number=maximum_number
        self.__count=count
    
    def increment(self):
        if self.__count >= self.__maximum_number:
            raise ValueError("Error, counter has reached its maximum_value",self.__maximum_number)
        else:
            self.__count+=1
        
    def reset(self):
        self.__count=0

    def counter_value(self):
        return self.__count
    
    def maximum_value(self):
        return self.__maximum_number
    
    def is_c_at_maximum(self):
        return self.__count==self.__maximum_number
    
    def __repr__(self):
        return f"counter(count={self.__count}, maximum={self.__maximum_number})"
    

```

```python
class Rectangle:
    def __init__(self, l,w, x,y):
        if l is None or w is None:
            raise ValueError("Please enter some value for the length and width")
        
        self.__l=l
        self.__w=w
        self.__x=x
        self.__y=y
        

    def area(self):
        return self.__l * self.__w
    
    def perimeter(self):
        return 2 * (self.__l+self.__w)
    
    def get_length(self):
        return self.__l
    
    def get_width(self):
        return self.__w
    
    
    
    def get_coordinates(self):
        return (self.__x,self.__y)
    


    
    
```

```python
#Implementing on Triangle
class Triangle(Shape):
    def __init__(self,a,b,c,x,y):
        super().__init__(x,y)
        # validating triangle
        # sum of two sides should always be greater than  to the third side
        if a+b>c and b+c>a and c+a>b:

            self.__a=a
            self.__b=b
            self.__c=c
        else:
            raise ValueError("Sum of the two sides of the triangle should always be greater than third side")
        

    def perimeter(self):
        return self.__a+self.__b+ self.__c
    
    def area(self):
        semi=self.perimeter()/2
        return math.sqrt(semi*(semi-self.__a)*(semi-self.__b)*(semi-self.__c))
    
    def get_a(self):
        return self.__a

    def get_b(self):
        return self.__b
    
    def get_c(self):
        return self.__c
    
    def _get_vertices(self):
        
        #first of all finding the point of the vertices
        v1=(self._x,self._y)
        v2=(self._x+self.__c,self._y) #here we consider c to be a horizontal line
            
        #for the third vertex
        #to be a triangle the third vertex can't be located just anywhere, hence 
        #to find its location that support the triangle, we are using the formula involving cos and sine

        cosA=(self.__b**2+self.__c**2-self.__a**2)/ (2*self.__b* self.__c)
        sinA=math.sqrt(1-cosA**2)

        v3=(self._x+self.__b* cosA, self._y+self.__b* sinA)

        return v1,v2,v3

    
    def perm_points(self,max_points=16):
        if max_points< 1 or max_points > 16:
            raise ValueError("Points must be 1 and 16")
        

        v1,v2,v3=self._get_vertices()
        vertices=[v1,v2,v3]

        #Now after finding the vertices we want to generate the points across the edges

        points=[]
        i=0

        while len(points)< max_points:
            start=vertices[i%3]
            end=vertices[(i+1)%3]

            move_size=(len(points)+1)/max_points

            x=start[0]+ move_size*(end[0]-start[0])
            y=start[1]+move_size*(end[1]-start[1])
        
            points.append((x,y))
            i+=1
        return points[:max_points]
    
    
    def test_points(self,x1,y1):
        v1,v2,v3=self._get_vertices()

        def calculated_area(v1,v2,v3):
            return abs(1/2* (v1[0]*(v2[1]-v3[1])+ v2[0]*(v3[1]-v1[1])+v3[0]*(v1[1]-v2[1])))
        
        tot_area= calculated_area(v1,v2,v3)
        area1=calculated_area(v1,v2,(x1,y1))
        area2=calculated_area(v2,v3,(x1,y1))
        area3=calculated_area(v3,v1,(x1,y1))
        return abs((area1+area2+area3)-tot_area) <1e-6

    



    
    def __repr__(self):
        return f"Triangle of side length {(self.__a,self.__b,self.__c)} , Coordinates{self._x,self._y})"
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)

    #got a division by zero error hence used LLM (chatgpt and it flagged line as a problem hence working on it)
    #it didnot handled the vertical line properly
    def line(self, x1, y1, x2, y2, **kargs):
        steps=max(abs(x2-x1),abs(y2-y1))

        if steps==0:
            self.set_pixel(y1,x1,**kargs)
            return

        for i in range(steps+1):
            t=i/steps
            x=int(x1+t*(x2-x1))
            y=int(y1+t*(y2-y1))
            self.set_pixel(y,x, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
#Implemeting Parent class
class Shape:
    def __init__(self,x,y):
        self._x=x
        self._y=y
        
    def get_coordinates(self):
        return(self.__x,self.__y)
    def area(self):
        raise NotImplementedError("Subclasses should implement area")
    
    def perimeter(self):
        raise NotImplementedError("subclasses should implemented perimeter")
        
    def perm_points(self):
        raise NotImplementedError("Subclasses should implement perm_points")
        
    def test_points(self,x1,y1):
        raise NotImplementedError("Subclass should Implement the Test")
    
    
    def overlapping_check(self,other,max_points=16):
        self_p=self.perm_points(max_points)
        other_p=other.perm_points(max_points)
        #checking if any of the shape's boundary are inside the other shape
        for ptts in self_p:
            if other.test_points(ptts[0],ptts[1]):
                return True
        
        for ptts in other_p:
            if self.test_points(ptts[0],ptts[1]):
                return True
        return False
        


        
    def paint(self,canvas,char="*"):
        raise NotImplementedError("Paint should be Implemented by Subclass")
```

```python
#implementing the Rectangle class
class Rectangle(Shape):
    def __init__(self,x,y,l,w):
        super().__init__(x,y)
        if l is None or w is None:
            raise ValueError("Please enter some value for the length and width")
        
        self.__l=l
        self.__w=w
        
    def area(self):
        return self.__l* self.__w
        
    def perimeter(self):
        return 2 * (self.__l + self.__w)
    
    def get_length(self):
        return self.__l
    
    def get_width(self):
        return self.__w
    
    def perm_points(self,max_points=16):
        if max_points <1 or max_points > 16:
            raise ValueError("Points must be between 1 and 16")
        
        # first finding the step size based on the max_points
        points=[]
        step=max(self.__l,self.__w)/ max_points

        i=0
        while len(points)< max_points:
            # top_edge
            points.append((self._x+self.__l-(i*step) % self.__l, self._y+self.__w))
            if len(points) >=max_points: break

            # bottom_edge
            points.append((self._x+(i*step) % self.__l,self._y))
            if len(points) >=max_points:break

            #right_edge
            points.append((self._x+self.__l,self._y+(i*step)%self.__w))
            if len(points) >=max_points: break

            #left_edge
            points.append((self._x,self._y+self.__w-(i*step)%self.__w))

            i+=1
        return points[:max_points]
    
    def test_points(self,x1,y1):
        return (self._x<=x1 <=self._x + self.__l) and (self._y<=y1<=self._y+ self.__w)
    
    def paint(self,canvas, char="*"):
        row=self._y
        col=self._x
        canvas.h_line(row,col,self.__l)
        canvas.h_line(row+self.__w,col,self.__l)
        canvas.v_line(row,col,self.__w)
        canvas.v_line(row,col+self.__l,self.__w)
    def __repr__(self):
        return f"Rectangle of length {self.__l}, Width {self.__w}, and Corner Coordinates: {(self._x,self._y)}"
    


        
```

```python
class CompoundShape(Shape):
    def __init__(self):
        super().__init__(0,0)
        self.shapes=list()

    def add_shape(self,shape):
        self.shapes.append(shape)
    
    def remove_shape(self,shape):
        self.shapes.remove(shape)

    def perimeter(self):
        return sum(shape.perimeter() for shape in self.shapes)
    
    def area(self):
        return sum(shape.area() for shape in self.shapes)
    
    def perm_points(self,max_points=16):
        ptts=[]
        for shape in self.shapes:
            ptts.extend(shape.perm_points(max_points))
        return ptts[:max_points]
    
    def test_points(self,x1,y1):
        return any(shape.test_point(x1,y1) for shape in self.shapes)
    
    def paint(self,canvas, char="*"):
        for shape in self.shapes:
            shape.paint(canvas,char)

    def __repr__(self):
        return f"CompoundShape with {len(self.shapes)} shapes in it"
```

```python
class RasterDrawing:
    def __init__(self,cavas):
        self.canvas=canvas
        self.shapes=[]
    def add_shape(self, shape):
        self.shapes.append(shape)
    
    def remove_shape(self, shape):
        self.shapes.remove(shape)
    
    def paint(self):
        self.canvas.clear_canvas()  
        for shape in self.shapes:
            shape.paint(self.canvas)
    
    def __str__(self):
        return f"RasterDrawing with {len(self.shapes)} shapes"
    
```
